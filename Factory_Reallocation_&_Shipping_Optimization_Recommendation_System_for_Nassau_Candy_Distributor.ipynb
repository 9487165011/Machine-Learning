```python
import numpy as np
from pulp import LpProblem, LpMinimize, LpVariable, lpSum

# Example data
factories = ["Factory_A", "Factory_B"]
warehouses = ["NY", "LA", "Chicago"]

capacity = {"Factory_A": 500, "Factory_B": 700}
demand = {"NY": 300, "LA": 400, "Chicago": 500}
shipping_cost = {
    ("Factory_A", "NY"): 2,
    ("Factory_A", "LA"): 5,
    ("Factory_A", "Chicago"): 3,
    ("Factory_B", "NY"): 4,
    ("Factory_B", "LA"): 2,
    ("Factory_B", "Chicago"): 6,
}

# Optimization model
model = LpProblem("Factory_Reallocation", LpMinimize)

# Decision variables
x = LpVariable.dicts("ship", (factories, warehouses), lowBound=0)

# Objective: minimize total shipping cost
model += lpSum([shipping_cost[(f, w)] * x[f][w] for f in factories for w in warehouses])

# Constraints
for f in factories:
    model += lpSum([x[f][w] for w in warehouses]) <= capacity[f]

for w in warehouses:
    model += lpSum([x[f][w] for f in factories]) >= demand[w]

# Solve
model.solve()

# Results
for f in factories:
    for w in warehouses:
        print(f"{f} -> {w}: {x[f][w].varValue}")
```
